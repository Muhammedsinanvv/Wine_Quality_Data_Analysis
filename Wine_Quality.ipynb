importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

reading the data set winequality.csv

```python
data=pd.read_csv('winequality.csv')
```

plotting a histogram of fixed acidity

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data, x= 'fixed acidity',color='orange', edgecolor='linen',alpha=0.5,bins=5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('count')
plt.show()
```

plotting the mean and median in the histogram

```python
print(round(data['fixed acidity'].mean(),2))
```

```python
data['fixed acidity'].median()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data, x= 'fixed acidity',color='orange', edgecolor='linen',alpha=0.5,bins=5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('count')
plt.vlines(data['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue',label='Mean')
plt.vlines(data['fixed acidity'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

plotting a histogram of volatile acidity

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='volatile acidity',color='green',edgecolor='linen',alpha=0.5,bins=5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('count')
plt.show()
```

plotting the distribution curve of the volatile acidity

```python
plt.figure(figsize=(11,6))
sns.displot(data['volatile acidity'],color='blue')
plt.title("Distribution of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.show()
```

```python
print(data['volatile acidity'].skew())
```

plotting the mean and median in the histogram

```python
print(data['volatile acidity'].mean())
```

```python
print(data['volatile acidity'].median())
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='volatile acidity',color='green',edgecolor='linen',alpha=0.5,bins=5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('count')
plt.vlines(data['volatile acidity'].mean(),ymin=0,ymax=4000,colors='blue',label='Mean')
plt.vlines(data['volatile acidity'].median(),ymin=0,ymax=4000,colors='red',label='Median')
plt.legend()
plt.show()
```

plotting the histogram of citric acidity with it's mean and median

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='citric acid',color='red',edgecolor='linen',alpha=0.5,bins=5)
plt.title("Histogram of Citric Acidity")
plt.xlabel('Citric Acidity')
plt.ylabel('count')
plt.vlines(data['citric acid'].mean(),ymin=0,ymax=4000, colors='blue',label='Mean')
plt.vlines(data['citric acid'].median(),ymin=0,ymax=4000, colors='red',label='Median')
plt.legend()
plt.show()
```

plotting the distribution curve of the citric acidity

```python
plt.figure(figsize=(11,6))
sns.displot(data['citric acid'],color='blue')
plt.title("Distribution of Citric Acid")
plt.xlabel('Citric Acid')
plt.ylabel('Density')
plt.show()
```

visualizing the wine's quality

```python
plt.figure(figsize=(11,6))
sns.countplot(x='quality',data=data)
plt.title("Count Plot of Wine Quality")
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
data['quality'].value_counts()
```

```python
print(data['quality'].value_counts().index[0])
```

final summary of the dataset

```python
rep_acid=pd.Series(index=['fixed acidity','volatile acidity','citric acid','quality'],
                   data=[data['fixed acidity'].mean(),data['volatile acidity'].mean(),
                         data['citric acid'].mean(),data['quality'].value_counts().index[0]])
```
